```
import urllib.request
url = "https://raw.githubusercontent.com/eresytter/lazada2019/main/lazada_train.csv"
urllib.request.urlretrieve(url, 'train.csv')
from nltk.corpus import stopwords
```

```
from collections import Counter
import csv
import re
import pandas as pd
import numpy as np
import itertools
```

```
df = pd.read_csv("train.csv", nrows=3000)
df.head()
```

```
df.shape # how many rows do we have?
```

```
df.sentiment.value_counts() # check how many pos reviews vs neg reviews
```

```
from sklearn.feature_extraction.text import TfidfVectorizer
```

```
vectorizer = TfidfVectorizer(max_features=1000)
vectors = vectorizer.fit_transform(df.reviewContent)
words_df = pd.DataFrame(vectors.toarray(), columns=vectorizer.get_feature_names())
words_df.head()
```

```
x = words_df
y = df.sentiment
```

```
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import LinearSVC
from sklearn.naive_bayes import MultinomialNB
```

```
%%time
# Create and train a logistic regression
logreg = LogisticRegression(C=1e9, solver='lbfgs', max_iter=1000)
logreg.fit(x, y)
```

```
%%time
# Create and train a random forest classifier
forest = RandomForestClassifier(n_estimators=50)
forest.fit(x, y)
```

```
%%time
# Create and train a linear support vector classifier (LinearSVC)
svc = LinearSVC()
svc.fit(x, y)
```

```
%%time
# Create and train a multinomial naive bayes classifier (MultinomialNB)
bayes = MultinomialNB()
bayes.fit(x, y)
```

```
pd.set_option('display.max_colwidth', 200)
url2 = "https://raw.githubusercontent.com/eresytter/lazada2019/main/lazada_test.csv"
urllib.request.urlretrieve(url2, 'test.csv')
test = pd.read_csv("test.csv", nrows=3000)
test.head()
```

```
test_vectors = vectorizer.transform(test.content)
test_words_df = pd.DataFrame(test_vectors.toarray(), columns=vectorizer.get_feature_names())
test_words_df.head()
```

```
test_words_df.shape
```

```
test['pred_logreg'] = logreg.predict(test_words_df)
```

```
# Logistic Regression prediction & probabilities
test['pred_logreg'] = logreg.predict(test_words_df)
test['pred_logreg_proba'] = logreg.predict_proba(test_words_df)[:,1]

# Random forest prediction & probabilities
test['pred_forest'] = forest.predict(test_words_df)
test['pred_forest_proba'] = forest.predict_proba(test_words_df)[:,1]

# SVC predicitons
test['pred_svc'] = svc.predict(test_words_df)

# Bayes predictions + probabilities
test['pred_bayes'] = bayes.predict(test_words_df)
test['pred_bayes_proba'] = bayes.predict_proba(test_words_df)[:,1]
```

```
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y)
```
